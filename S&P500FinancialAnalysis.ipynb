Download Pricing Data for all the companies of S&P 500 Stock Market Index using Yahoo Finance 

```python
import pandas as pd
import yfinance as yf
```

```python
snp500url= "https://en.wikipedia.org/wiki/List_of_S%26P_500_companies"
```

```python
datatable = pd.read_html(snp500url)
```

```python
#Lets Get this array of symbols in a list.
sp_tickers= datatable[0]['Symbol'].tolist()
```

```python
yf.download('MMM', start = '2024-01-01', end = '2024-01-31')
```

```python
sp_prices = yf.download(sp_tickers, start='2024-01-01', end='2024-01-10')['Adj Close']
```

```python
#we have to get these tickers name updated in the list hence.. For Loop

for i in range(len(sp_tickers)):
    if sp_tickers[i]=='BF.B':
        sp_tickers[i] = 'BF-B'
    elif sp_tickers[i] == 'BRK.B':
        sp_tickers[i] = 'BRK-B'
```

```python
'BRK-B' in sp_tickers
'BF-B' in sp_tickers
```

```python
sp_prices = yf.download(sp_tickers, start='2024-01-01', end='2024-01-10')['Adj Close']
```
